Mathematical operations

```python
a = np.array([2,4,6])
b = np.array([1,3,5])
c =np.add(a,b)
print(c)
d = np.subtract(a,b)
print(d)
e = np.multiply(a,b)
print(e)
f = np.divide(a,b)
print(f)
g = np.dot(a,b)
print(g)
```
